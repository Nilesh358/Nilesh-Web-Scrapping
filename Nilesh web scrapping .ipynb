```python
# Import libraries

import requests
import bs4
```

```python
# Make a request using request library

res=requests.get("https://en.wikipedia.org/wiki/Charles_Lane_Poole")
```

```python
res.text
```

```python
#To change the string into HTML+CSS code, we are going to use BeautifulSoup library

mytoc=bs4.BeautifulSoup(res.text,"lxml")
```

```python
mytoc
```

```python
mytoc.select(".toctext")
# It gives the list of all the table of content
```

```python
first_topic=mytoc.select(".toctext")[0]

first_topic
```

```python
# To retrive data from the above code
first_topic.text
```

```python
# To print all the text from mytoc, I need to use loop
for item in mytoc.select(".toctext"):
    
    print(item.text)
```
